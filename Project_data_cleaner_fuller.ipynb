```python
#Процедура отчистки данных по времени
def X_cleaner(path,data_lims,errors_lims,path_out):
    m = len(path)
    for i in range(0,m):
        data = pd.read_csv(path[i])
        
        
        
        data = data.loc[data['RIGTIME'] >= data_lims[i][0]]
        data = data.loc[data['RIGTIME'] < data_lims[i][1]]
        
        data.to_csv(path_out[i])
        del data
```

```python
# список параметров которые есть во всех датасетах
def general_list_of_cols(All_cols,path):
    for i in range(0,len(path)):
        data = pd.read_csv(path[i])
        for j in All_cols:
            if j not in data.columns:
                All_cols.remove(j)
        del data
    return All_cols
```

```python
# список параметров которые есть во всех датасетах c пропусками
def spaces_of_cols(General_cols,path_cl):
    for i in range(0,len(path_cl)):
        data = pd.read_csv(path_cl[i])
        for col in data.columns:
            pct_missing = round(np.mean(data[col].isnull())*100,2)
            for j in General_cols:
                if col == j:
                    d_space.iloc[i][j] = pct_missing
            
        del data
    return d_space
```

```python
#Процедура заполнения очистки от ненужных параметров, заполнения пропусков в значимых и разметка.
def full(path,par,par2):
    for i in range(0,len(path)):
        data = pd.read_csv(path[i])
        data = data[par]
        datatime = data['RIGTIME'].tolist()
        flag = np.zeros((len(datatime),), dtype=int)
        
        
        for j in range(0,len(datatime)):
            if datatime[j] >= errors_limits[i][0] and datatime[j] <= errors_limits[i][1]:
                flag[j] = 1       
        ## зануление параметров в случае если случайно 100% пропусков
        for col in par2:
            pct_missing = round(np.mean(data[col].isnull())*100)
            if pct_missing == 100:
                data.loc[:, col]=0
        
        
        
        data = data.ffill()
        data = data.bfill()
        data.insert(len(par), 'flag',flag)
        data.to_csv(path[i])
        del data
        
```
