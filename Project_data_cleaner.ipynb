```python
def general_list_of_cols(All_cols,dfraims):
    All_cols = list(All_cols)
    for i in dfraims:
        for j in All_cols:
            if j not in i.columns:
                All_cols.remove(j)
    print("Общий список параметров имеет длинну {}".format(len(All_cols))) 
```

```python
def cols_filter(List_useful,dfraims,eps):
    # eps - limit of permissible omissions
    
    ## Поиск минимального списка параметров
    for i in dfraims:
        for col in i.columns:           
            pct_missing = np.mean(i[col].isnull())
            if pct_missing > eps:
                if col in List_useful:
                    List_useful.remove(col)
```

```python
general_list_of_cols(All_cols,data0)

List_using_cols = All_cols
All_cols = list(All_cols)
List_using_cols = list(List_using_cols)
eps = 0.5 ## limit of permissible omissions

cols_filter(List_using_cols, data0 ,eps)
print(len(List_using_cols))
```
